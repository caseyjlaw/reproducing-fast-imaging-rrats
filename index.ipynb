# Basic visualization of candidates produced by rtpipe
## In support of publication titled "Fast Imaging of RRATs"
## Casey Law

```python
import rtpipe.parsecands as pc
import numpy as np
import pylab as pl
import logging

logger = logging.getLogger()
for hdlr in logger.handlers:
    logger.removeHandler(hdlr)
logging.basicConfig()

%matplotlib inline
```

```python
loc, prop, d = pc.read_candidates('cands_12A-336_sb9667618_1b.56040.87127945602_merge.pkl',
                                  returnstate=True, snrmin=8.)
```

```python
# set properties for plotting
times = pc.int2mjd(d, loc)
dms = np.array(d['dmarr'])[loc[:,3]]
snrs = prop[:,0]
l1 = prop[:,2]
m1 = prop[:,3]
```

```python
pl.scatter(times, dms, s=snrs**2, facecolor='none', edgecolor='red')
```

```python
pl.scatter(l1, m1, s=snrs**2, facecolor='none', edgecolor='red')
```
